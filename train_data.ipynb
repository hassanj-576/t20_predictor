```python
import pandas as pd
```

```python
df = pd.read_csv("./data.csv")
```

```python
def is_winner(row):
    if type(row['result']) ==str:
        if row['result'] in row['team_2']:
            return 1
        return 0
    else:
        return 0
```

```python
# Method to calculate if the team batting second won or not for IPL
def ipl_is_winner(row):
    if str(row['batting_team'])== str(row['winner']):
        return 1
    else:
        return 0
    return 0
```

```python
# Method to calculate if a player was dismissed on this ball or not
def is_out(row):
    if type(row['player_dismissed']) == str:
        return 1
    else:
        return 0
```

# Winner Prediction Data

```python
# Splitting on Innings
first_inning = df[df['inning']==1]
second_inning = df[df['inning']==2]
```

```python
#Calculating Target
total_sum = first_inning.groupby(["psl_year","match_number"]).agg(
   target= ("runs","sum")
)
```

```python
total_sum = total_sum.reset_index()
```

```python
total_sum
```

```python
# Merging the DF
new_df = pd.merge(total_sum, second_inning,  how='left', left_on=['psl_year','match_number'], right_on = ['psl_year','match_number'])
                                                                                                          
                                                                                                          
```

```python
# Calculating if Second Inning team won or not 
new_df['won'] = new_df.apply(is_winner, axis=1)
```

```python
# Calculating balls left
new_df['balls_left'] = 120-(((new_df['over']-1)*6)+new_df['ball'])
# Calculating balls left
new_df['wickets_left'] = 10-new_df['wickets']
```

```python
# Creating the Final Dataset
final_df = new_df[['over','ball','total_runs','wickets_left','target','balls_left','won']]
```

```python
final_df
```

```python
final_df.to_csv("winner_prediction_data.csv")
```

# Training Winning Prediction

```python
import pandas as pd
import sklearn as sk
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import time
import joblib
```

```python
prediction_df = pd.read_csv("winner_prediction_data.csv")
```

```python
prediction_df['runs_left']= prediction_df['target']-prediction_df['total_runs']
```

```python
prediction_df.head(10)
```

```python
new_cols =['wickets_left','balls_left','runs_left','won']
```

```python
print(prediction_df.shape)
prediction_df=prediction_df.dropna()
print (prediction_df.shape)
```

```python
psl_df=prediction_df[new_cols]
psl_df
```

```python
ipl_df = pd.read_csv("ipl_formated.csv")
```

```python
ipl_df
```

```python
full_df = pd.concat([psl_df,ipl_df], ignore_index=True)
```

```python
full_df
```

```python
# X and Y Split
y = psl_df.iloc[:,3]
x = psl_df.iloc[:,:3]
```

```python
# Train and Test Split
X_train, X_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42)
```

```python
# Training Model
start = time.time()
RF = RandomForestClassifier(n_estimators=10000, max_depth=6, random_state=0)
RF.fit(X_train, y_train)
print(f"Time Taken to Train {time.time()-start} seconds")
```

```python
y_pred_test = RF.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_test)
```

```python
current= {
    "wickets_left":10,
    "balls_left" :10,
    "runs_left":1
}
current_df = pd.DataFrame(current,index=[0])
RF.predict_proba(current_df)
```

```python
# Saving Model
joblib.dump(RF, "./prediction_model.joblib")
```
