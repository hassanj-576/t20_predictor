```python
import pandas as pd
```

# Data from Kaggle
https://www.kaggle.com/datasets/vora1011/ipl-2008-to-2021-all-match-dataset

```python
# Method to calculate if the team batting second won or not for IPL
def ipl_is_winner(row):
    if str(row['BattingTeam'])== str(row['WinningTeam']):
        return 1
    else:
        return 0
    return 0
```

```python
score = pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
```

```python
score.head(1)
```

```python
matches = pd.read_csv("IPL_Matches_2008_2022.csv")
```

```python
matches.head(2)
```

```python
ipl_match_winner = matches[['ID','WinningTeam']]
```

```python
ipl_match_winner
```

```python
ipl_df = pd.merge(ipl_match_winner, score,  how='left', left_on=['ID'], right_on = ['ID'])
```

```python
first_inning =  ipl_df[ipl_df['innings']==1]
```

```python
first_inning_total = first_inning.groupby("ID").agg(
target = ("total_run","sum")
)
```

```python
first_inning_total
```

```python
second_inning = ipl_df[ipl_df['innings']==2]
```

```python
second_inning.head(3)
```

```python
second_inning = second_inning[['ID','overs','BattingTeam','ballnumber','total_run','isWicketDelivery','WinningTeam']]
```

```python
second_inning
```

```python
second_inning['total'] = second_inning.groupby("ID").agg(
    total=('total_run','cumsum')
)
```

```python
second_inning.head(30)
```

```python
second_inning['wickets'] =second_inning.groupby("ID").agg(
    total=('isWicketDelivery','cumsum')
)
```

```python
first_inning_total= first_inning_total.reset_index()
```

```python
first_inning_total
```

```python
merged_df = pd.merge(first_inning_total, second_inning,  how='left', left_on=['ID'], right_on = ['ID'])
```

```python
merged_df
```

```python
merged_df['runs_left']= merged_df['target'] -merged_df['total']
```

```python
merged_df
```

```python
merged_df['balls_left'] = 120 -(((merged_df['overs']-1)*6)+merged_df['ballnumber'])
```

```python
merged_df
```

```python
merged_df['won'] =merged_df.apply(ipl_is_winner, axis=1)
```

```python
merged_df['wickets_left'] = 10 - merged_df['wickets']
```

```python
merged_df
```

```python
final_df_ipl = merged_df[['wickets_left','balls_left','runs_left','won']]
```

```python
final_df_ipl
```

```python
final_df_ipl.to_csv("ipl_formated.csv",index=False)
```
